On fournit la fonction `derivee_vectorielle`.

```python
def derivee_vectorielle(M):
    r"""
    Renvoie la dérivée d'une fonction d'une variable réelle
    à valeurs vectorielles.

    EXAMPLES:

    Dans le plan, la cycloïde standard::

        sage: x(t) = t - sin(t)
        sage: y(t) = 1 - cos(t)
        sage: M(t) = (x(t), y(t))
        sage: M
        t |--> (t - sin(t), -cos(t) + 1)
        sage: M1 = derivee_vectorielle(M)
        sage: M1
        t |--> (-cos(t) + 1, sin(t))
    """
    t = SR.var('t')
    vt =  diff(M(t), t)
    v(t) = vt
    return v
```

**Repère de Frenet.**

Pour chaque courbe dans l'espace, on peut définir un repère mobile adapté au parcours de la courbe.
On l'appelle **repère de Frenet**.

Il comporte, en chaque point, trois vecteurs, notés $T$ pour tangent \(vecteur tangent unitaire\), $N$ pour normal \(vecteur normal unitaire principal\), $B$ pour binormal \(vecteur unitaire normal aux deux précédents\).

Pour une courbe parcourue à vitesse 1, on définit ces trois vecteurs comme suit:

- $T$ est égal au vecteur vitesse divisé par sa norme \(qui est la vitesse\)
- $N$ est égal au vecteur accélération divisé par sa norme \(qui est la courbure\)
- $B$ est égal au produit vectoriel de $T$ et $N$



Voici un calcul du repère de Frenet pour la courbe que nous étudions ici.
On applique des fonctions de simplification aux coordonnées.

```python
T(t) = M1(t)
```

```python
N(t) = M2(t) / κ(t)
```

```python
N
```

```python
N(t) = N(t).apply_map(lambda expr: expr.simplify_real().simplify_trig())
```

```python
N
```

```python
B(t) = T(t).cross_product(N(t))
```

```python
B
```

```python
B(t) = B(t).apply_map(lambda expr: expr.simplify_real().simplify_trig())
```

```python
B
```
