# Le folium de Descartes

Le folium de Descartes de paramètre $a$ admet pour équation

$x^3 + y^3 - 3 a x y = 0$.

En regardant son intersection avec la droite d'équation $y = tx$ on obtient le paramétrage:

$
M(t) = \left[ \begin{matrix}\frac{3at}{1 + t^3} \\ \frac{3at^2}{1 + t^3} \\ \end{matrix} \right].
$

En vous inspirant de l'étude faite dans `exemple.ipynb`, faire une étude et un tracé avec Sage.



```python
a = 2
x(t) = (3*a*t)/(1+t^3)
y(t) = (3*a*(t^2))/(1+t^3)
```

```python
M(t) = (x(t),y(t))
```

```python
A = parametric_plot(M, (-18., -1.01), color= 'red')  # de -18. à -1.01 au lieu de -oo à -1
B = parametric_plot(M, (-0.99, 18), color= 'blue')   # de -0.99 à 18 au lieu de -1 à +oo
```

```python
courbe = A + B
courbe.show(xmin=-20, xmax=20, ymin=-15, ymax=15)
```
