## Quelques courbes classiques

### Graphes

Soit $f:I\to\mathbb R$ une fonction réelle de variable réelle, continue, définie sur un intervalle $I$. Alors, le __graphe__ de la function, c'est-à-dire l'ensemble $G=\{\,(x,f(x))\;|\;x\in I\,\}$, est une courbe.

Pour tracer le graphe d'une fonction dans Sage, on utilise la commande `plot`, dont vous pouvez retrouver la documentation ici:

https://doc.sagemath.org/html/en/reference/plotting/sage/plot/plot.html#sage.plot.plot.plot

La commande `plot` de Sage attend de vous deux paramètres indispensables:

- la fonction dont on souhaite tracer le graphe
- l'intervalle borné au dessus duquel on souhaite tracer le graphe



Par exemple, pour tracer le graphe du polynôme cubique $x^3 + 2\,x + 1$ sur un intervalle $[a, b]$, on exécute la commande `plot(x^3 + 2*x + 1, (x, a, b))` en ayant préalablement défini `a` et `b`, ou en utilisant des valeurs numériques à leur place. Par exemple:

```python
x = SR.var('x')
a = -2
b = 2
plot(x^3 + 2*x + 1, (x, a, b))
```

ou bien

```python
x = SR.var('x')
plot(x^3 + 2*x + 1, (x, -3, 3))
```

ou

```python
x = SR.var('x')
a, b = -3, 2
plot(x^3 + 2*x + 1, (x, a, b))
```

Dans les cours de calcul vous avez vu le concept de dérivée et son interprétation en tant que direction tangente au graphe.

Dans certaines circonstances, on aura envie de tracer un vecteur tangent, ou une droite tangente, au graphe d'une fonction.

Pour cela, on utilisera les mots clés `vector` ou `line`.

À titre d'exemple, prenons la fonction $\operatorname{Ai}$ d'Airy, définie comme suit:
$$Ai(x)=\frac1π∫^∞_0\cos\left(\frac13\,t^3+x\,t\right)dt$$
Il s'agit d'une des fonctions d'Airy, qui sont déjà prédéfinies dans Sage et qui résolvent l'équation différentielle $y″ = x \, y$.

La fonction $\operatorname{Ai}$ apparaît dans Sage comme `airy_ai`.

Vous retrouverez sa documentation ici:

https://doc.sagemath.org/html/en/reference/functions/sage/functions/airy.html#sage.functions.airy.airy_ai

Elle peut faire le calcul exact ou approché:

````python
airy_ai(0)
````

````python
airy_ai(0.0)
````

Rappellons que pour une fonction de classe $\mathcal C^2$, comme c'est le cas ici, les points d'inflexion sont ceux ou la dérivée seconde s'annule et change de signe.

Comme $\operatorname{Ai}$ satisfait l'équation $y″ = x \, y$ et qu'elle est positive en $0$, non seulement sa dérivée seconde s'annule en $0$ mais elle change aussi de signe. Bingo, c'est un point d'inflexion.

Traçons le graphe de $\operatorname{Ai}$ avec ce point d'inflexion et le vecteur tangent $(0, \operatorname{Ai}'(0))$ au point d'inflexion.
Sage a une fonction prédéfinie pour la première dérivée de $\operatorname{Ai}$, à savoir `airy_ai_prime`,
mais si elle n'existait pas, on aurait pu la calculer avec `diff`.

````python
P = vector((0, airy_ai(0)))
v = vector((1, airy_ai_prime(0)))
graphe = plot(airy_ai(x), (x, -10, 5))
inflexion = point([P], color='black')
vecteur_tangent = v.plot(start=P, thickness=1, color='black')
graphe + inflexion + vecteur_tangent
````

Réalisons maintenant le tracé avec un segment de la droite tangente passant par ce point d'inflexion - cela permet de mieux apprécier la nature du point d'inflexion.

````python
P = vector((0, airy_ai(0)))
v = vector((1, airy_ai_prime(0)))
droite_tangente = line([P - v, P + 3*v], thickness=1, color='black', ticks=[[],[]])
graphe + inflexion + droite_tangente
````

### Courbes paramétrées (dans le plan)



Une __courbe paramétrée__ (plane) est la donnée d'un couple ordonné de fonctions réelles de variable réelle, continues, définies sur un même intervalle réel.

Le __tracé d'une courbe__ paramétrée $$\left(\begin{array}{rcrcl}x&:& I&\to& \mathbb R\\&&t&\mapsto&x(t)\end{array}, \begin{array}{rcrcl}y&:& I&\to& \mathbb R\\&&t&\mapsto&y(t)\end{array}\right)$$
est l'ensemble des points de $\mathbb R^2$ parcourus par ses images, à savoir
$$\left\{\, \begin{pmatrix}x(t) \\ y(t) \\ \end{pmatrix} \; | \; t ∈ I \,\right\}$$
Ce tracé est aussi appelé __courbe__.

Pour tracer des courbes paramétrées dans Sage, on utilise `parametric_plot`. 

Par exemple, pour tracer une famille d'ellipses qui s'allongent au fur et à mesure tout en conservant l'aire, on peut utiliser:

````python
t = SR.var('t')
G = Graphics()
for i in srange(1, 7, 3/4):
    a = i
    b = 1/i
    G += parametric_plot((a*sin(t), b*cos(t)), (t, 0, 2*π), color=hue((i - 1)/7))
show(G, axes=False, aspect_ratio=1, figsize=12)
````

Ou en condensé:

````python
t = SR.var('t')
G = sum(parametric_plot((i * sin(t), 1/i * cos(t)), (t, 0, 2*π), color=hue((i - 1)/7))
        for i in srange(1, 7, 3/4))
G.show(axes=False, aspect_ratio=1, figsize=12)
````
