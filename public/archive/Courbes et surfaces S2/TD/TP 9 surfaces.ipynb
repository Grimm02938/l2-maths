# Surfaces

Quelques surfaces paramétrées:

- sphère: $(\cos u \cos v, \sin u \cos v, \sin v)$ pour $(u, v) \in [0, 2 \pi]^2$
- selle de singe: $(u, v, u^3 - 3 u v^2)$ pour $(u, v) \in \mathbb{R}^2$

```python
def sphere(r=1):
    x, y = var('x y')
    a = cos(x) * cos(y)
    b = sin(x) * cos(y)
    c = sin(y)
    return a, b, c, x, y

def selle_de_singe():
    x, y = var('x y')
    a = x
    b = y
    c = x^3 - 3 * x * y^2
    return a, b, c, x, y

def tangent_plane_normale(surface, point):
    a, b, c, x, y = surface
    x0, y0 = point
    a_dx = diff(a, x)
    a_dy = diff(a, y)
    b_dx = diff(b, x)
    b_dy = diff(b, y)
    c_dx = diff(c, x)
    c_dy = diff(c, y)
    tangente_x = vector([a_dx.subs({x: x0, y: y0}), b_dx.subs({x: x0, y: y0}), c_dx.subs({x: x0, y: y0})])
    tangente_y = vector([a_dy.subs({x: x0, y: y0}), b_dy.subs({x: x0, y: y0}), c_dy.subs({x: x0, y: y0})])
    normale = tangente_x.cross_product(tangente_y)
    normale_unitaire = normale / normale.norm()
    return tangente_x, tangente_y, normale_unitaire

S = sphere()

selle_singe = selle_de_singe()

tau = 2 * pi

plot_S = parametric_plot3d((S[0], S[1], S[2]), (0, tau), (0, tau), color = 'purple', opacity = 0.6)

point_sphere = (pi/4, pi/4)

tangente_x_S, tangente_y_S, normale_S = tangent_plane_normale(S, point_sphere)

normale_vect_plot = arrow3d((cos(point_sphere[0])*cos(point_sphere[1]), 
                              sin(point_sphere[0])*cos(point_sphere[1]), 
                              sin(point_sphere[1])),
                             (cos(point_sphere[0])*cos(point_sphere[1]) + normale_S[0],
                              sin(point_sphere[0])*cos(point_sphere[1]) + normale_S[1],
                              sin(point_sphere[1]) + normale_S[2]), color = 'green')

plot_selle_singe = parametric_plot3d((selle_singe[0], selle_singe[1], selle_singe[2]), (-2, 2), (-2, 2), color = 'brown', opacity = 0.6)

point_singe = (1, 1)

tangent_x_singe, tangent_y_singe, normale_singe = tangent_plane_normale(selle_singe, point_singe)


normale_vect_plot_singe = arrow3d((point_singe[0], point_singe[1], point_singe[0]^3 - 3 * point_singe[0] * point_singe[1]^2),
                                    (point_singe[0] + normale_singe[0], 
                                     point_singe[1] + normale_singe[1],
                                     point_singe[0]^3 - 3 * point_singe[0] * point_singe[1]^2 + normale_singe[2]), color = 'blue')

show(plot_S + normale_vect_plot)
show(plot_selle_singe + normale_vect_plot_singe)
```
