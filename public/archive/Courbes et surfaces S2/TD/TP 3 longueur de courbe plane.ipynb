> # Longueur d'une courbe plane

> Si $M : [a, b] \to \mathbb{R}^2$ est une courbe paramétrée régulière, sa longueur est:
>
> > $
> L_{[a, b]}(M) = \int_a^b \left\| M'(t) \right\| dt
> $


> Dans SageMath, si on a défini comme dans les TP précédents
>
> - une courbe paramétrée `M`
> - la fonction `derivee_vectorielle`
>
> on peut définir la dérivée $M'$ \(nommée ici `M1`\), puis la longueur de $M$, comme ceci:
>
> ```py
> M1 = derivee_vectorielle(M)
> t = SR.var('t')
> L = integral(M1(t).norm(), t, a, b)
> ```
>
> Il s'agit d'un calcul exact de l'intégrale.
>
> Si ce calcul n'aboutit pas, l'intégrale est renvoyée telle quelle, non évaluée.
>
> Pour obtenir une valeur numérique de la longueur, deux méthodes existent.
>
> La première consiste à évaluer numériquement le résultat du calcul exact.
>
> ```py
> L_num_a = L.norm()
> ```
>
> La deuxième consiste à opter dès le départ pour un calculer numérique,
> plutot qu'exact, de l'intégrale.
>
> ```py
> L_num_b = numerical_integral(M1(t).norm(), a, b)
> ```
>
> Dans les exercices suivants, effectuer le calcul exact puis numérique,
> et comparer les valeurs numériques obtenues par les deux méthodes.

> **Exercice.** Calculer la longueur de l'astroïde de grand rayon 1.
>
> Rappel: un paramétrage de cette astroïde est:
>
> > $
> M(t) = \begin{bmatrix} \cos^3(t) \\ \sin^3(t) \\ \end{bmatrix}
> $
>
> Suggestion: commencer par calculer la longueur d'une arche d'astroïde, correspondant à l'intervalle de temps $[0, \pi/2]$.



````python
x(t)= cos(t)**3
y(t)= sin(t)**3
M(t)=(x(t), y(t))

M_astroide(t) = diff(M(t), t)

t = SR.var('t')

L_astroide = 4*integral(M_astroide(t).norm(),t, 0, pi/2)
L_astroide_num_a = L_astroide.n()

courbe = parametric_plot(M, (0, 2*pi), color='red')
courbe.show()

L_astroide_num_a
````

````python
M_Cay_Derivee = derivee_vectorielle(M)

L_Cay = 2*integral(M_Cay_Derivee(t).norm(),t, 0, pi/2)
L_Cay_num_a = L_Cay.n()
L_Cay, L_Cay_num_a
````
