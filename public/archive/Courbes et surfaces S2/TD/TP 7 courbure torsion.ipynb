> # Courbure et torsion


> On va tracer des courbes dans l'espace et calculer leur courbure et leur torsion.



> Le tracé des courbes s'effectue avec `parametric_plot` comme pour les courbes planes.



> ## Courbe des puissances

> Une courbe célèbre, qui se généralise à toute dimension, est la courbe des puissances.



```python
P(t) = (t, t^2, t^3)
```

```python
P(t) = (t,t^2,t^3)
```

```python
t = var('t')

def P(t):
    return vector([t, t^2, t^3])

t_min, t_max = -2, 2

plot_curve = parametric_plot((P(t)[0], P(t)[1], P(t)[2]), (t, t_min, t_max), color='blue')

def vitess_vect(t):
    return vector([diff(P(t)[i], t) for i in range(3)])

def tangent_vect(t_val):
    return vitess_vect(t).subs(t=t_val)

points = [-2, -1, 0, 1, 2]
sing_points = []

for t_val in points:
    v = vitess_vect(t).subs(t=t_val)
    if v.norm() == 0:
        sing_points.append(t_val)

tangent_vects = sum([arrow3d(P(t_val), P(t_val) + tangent_vect(t_val), color='red', width=0.05) for t_val in points if t_val not in sing_points])

def frenet_repere(t_val):
    T = vitess_vect(t).subs(t=t_val).normalized()
    a = vector([diff(T[i], t).subs(t=t_val) for i in range(3)]).simplify_full()
    N = a.normalized() if a.norm() != 0 else vector([0, 0, 0])
    B = T.cross_product(N)
    return T, N, B

frenet_reperes = sum([arrow3d(P(t_val), P(t_val) + frenet_repere(t_val)[0], color='green', width=0.05) + 
                     arrow3d(P(t_val), P(t_val) + frenet_repere(t_val)[1], color='orange', width=0.05) +  
                     arrow3d(P(t_val), P(t_val) + frenet_repere(t_val)[2], color='purple', width=0.05)  
                     for t_val in points if t_val not in sing_points])

show(plot_curve + tangent_vects + frenet_reperes)

def curve(t):
    v = vitess_vect(t)
    a = vector([diff(v[i], t) for i in range(3)]).simplify_full()
    return (v.cross_product(a).norm() / v.norm()^3).simplify_full()

def torsion(t):
    v = vitess_vect(t)
    a = vector([diff(v[i], t) for i in range(3)]).simplify_full()
    b = v.cross_product(a)
    b_deriv = vector([diff(b[i], t) for i in range(3)]).simplify_full()
    return (b.dot_product(b_deriv) / b.norm()^2).simplify_full()

curve_function = curve(t).simplify_full()
torsion_function = torsion(t).simplify_full()
curve_function, torsion_function
```

> ## Hélice au-dessus d'une courbe

> Étant donnée une courbe dans le plan, on définit l'hélice basée sur cette courbe en ajoutant une composante $z$ qui est une fonction affine de la longueur parcourue.

> **Exercice.**
>
> - Tracez une hélice à base circulaire.
> - Tracez le repère de Frenet en quelques points.

```python
t = var('t')

a, b = 1, 0.5  

def helice(t):
    return vector([a * cos(t), a * sin(t), b * t])

t_min, t_max = 0, 4 * pi

plot_helice = parametric_plot((helice(t)[0], helice(t)[1], helice(t)[2]), (t, t_min, t_max), color = 'blue')

def vitess_vect(t):
    return vector([diff(helice(t)[i], t) for i in range(3)])

def frenet_repere(t_val):
    T = vitess_vect(t).subs(t=t_val).normalized()  
    a = vector([diff(T[i], t).subs(t=t_val) for i in range(3)]).simplify_full()
    N = a.normalized() if a.norm() != 0 else vector([0, 0, 0]) 
    B = T.cross_product(N) 
    return T, N, B

points = [0, pi/2, pi, 3*pi/2, 2*pi, 5*pi/2, 3*pi, 7*pi/2, 4*pi]

frenet_reperes = sum([arrow3d(helice(t_val), helice(t_val) + frenet_repere(t_val)[0], color = 'black', width=0.05) +  
                     arrow3d(helice(t_val), helice(t_val) + frenet_repere(t_val)[1], color = 'green', width=0.05) + 
                     arrow3d(helice(t_val), helice(t_val) + frenet_repere(t_val)[2], color = 'red', width=0.05) 
                     for t_val in points])

show(plot_helice + frenet_reperes)
```

> ## Noeud de huit

> Voici un paramétrage censé produire un noeud de huit:
>
> - $x(t) = 10(\cos(t) + \cos(3 t)) + \cos(2 t) + \cos(4 t)$
> - $y(t) = 6 \sin(t) + 10 \sin(3 t)$
> - $z(t) = 4 \sin(3 t) \sin(5 t / 2) + 4 \sin(4 t) - 2 \sin(6 t)$
>
> **Exercice.**
>
> - Quelle est la période de cette courbe paramétrée?
> - Tracez la courbe. Est-ce un noeud de huit?
> - Tracez le repère de Frenet en quelques points.


```python
t = var('t')


def a(t):
    return 10 * (cos(t) + cos(3*t)) + cos(2*t) + cos(4*t)

def b(t):
    return 6 * sin(t) + 10 * sin(3*t)

def c(t):
    return 4 * sin(3*t) * sin(5*t/2) + 4 * sin(4*t) - 2 * sin(6*t)

def noeud(t):
    return vector([a(t), b(t), c(t)])

t_min, t_max = 0, 2 * pi
plot_noeud = parametric_plot((a(t), b(t), c(t)), (t, t_min, t_max), color = 'red')


def vitess_vect(t):
    return vector([diff(a(t), t), diff(b(t), t), diff(c(t), t)])

def frenet_repere(t_val):
    T = vitess_vect(t).subs(t=t_val).normalized()
    a = vector([diff(T[i], t).subs(t=t_val) for i in range(3)]).simplify_full()
    N = a.normalized() if a.norm() != 0 else vector([0, 0, 0])
    B = T.cross_product(N)
    return T, N, B

points = [0, pi/2, pi, 3*pi/2, 2*pi]
frenet_reperes = sum([arrow3d(noeud(t_val), noeud(t_val) + frenet_repere(t_val)[0], color = 'red', width=0.05) +  
                     arrow3d(noeud(t_val), noeud(t_val) + frenet_repere(t_val)[1], color = 'brown', width=0.05) +  
                     arrow3d(noeud(t_val), noeud(t_val) + frenet_repere(t_val)[2], color = 'black', width=0.05)
                     for t_val in points])

show(plot_noeud + frenet_reperes)
```

> ## Tube autour d'une courbe

> ### Tracé de surface paramétrée

> Pour tracer une surface paramétrée, on utilise `parametric_plot3d` qui prend en argument
>
> - soit une fonction de deux variables à valeurs vectorielles: la fonction $S: (u, v) \mapsto S(u, v)$
> - soit trois fonctions de deux variables à valeurs réelles: les fonctions
>   - abscisse $x : (u, v) \mapsto x(u, v)$
>   - ordonnée $y : (u, v) \mapsto y(u, v)$
>   - hauteur $z : (u, v) \mapsto z(u, v)$



```python
x, y = var('x y')

def a(x, y):
    return x

def b(x, y):
    return y

def c(x, y):
    return x^3 - 3*x*y^2

plot_selle = parametric_plot3d((a(x, y), b(x, y), c(x, y)), (x, -2, 2), (y, -2, 2), color = 'yellow', opacity = 0.6, mesh = True)

show(plot_selle)
```

> Remarque: la fonction `parametric_plot3d` accepte des arguments optionnels, comme `mesh=True` qui permet de montrer un maillage ou «structure en fil de fer» sur la surface.

> ### Paramétrage d'un tube le long d'une courbe

> Si $M$ est une courbe birégulière dans l'espace,
> pour obtenir un tube de rayon $r$ autour de $M$,
>
> - on utilise le repère de Frenet $(M, (T, N, B))$
> - on définit $S(t, \theta) = M(t) + r \cos(\theta) N(t) + r \sin(\theta) B(t)$

> **Exercice.**
>
> - Tracez un tube le long de chacune des courbes précédentes.

```python
x, y = var('x y')

def a(x, y):
    return cos(x) * cos(y)

def b(x, y):
    return sin(x) * cos(y)

def c(x, y):
    return sin(y)

def sphere_courbe(x, y):
    return vector([a(x, y), b(x, y), c(x, y)])

def vitess_vect(x, y):
    return vector([diff(a(x, y), x), diff(b(x, y), x), diff(c(x, y), x)])

def frenet_repere(x, y):
    T = vitess_vect(x, y).normalized()  
    a = vector([diff(T[i], u) for i in range(3)]).simplify_full()
    N = a.normalized() if a.norm() != 0 else vector([0, 0, 0])  
    B = T.cross_product(N)  
    return T, N, B

m = 0.2
teta = var('teta')

def tube_surface(x, teta):
    T, N, B = frenet_repere(x, 0)  
    return sphere_courbe(x, 0) + m * cos(teta) * N + m * sin(teta) * B

plot_tube_sphere = parametric_plot3d((tube_surface(x, teta)[0], tube_surface(x, teta)[1], tube_surface(x, teta)[2]), 
                                     (x, 0, 2*pi), (teta, 0, 2*pi), color = 'pink', opacity = 0.6, mesh = True)

show(plot_tube_sphere)
```

```python
u, v = var('u v')

def x(u, v):
    return u

def y(u, v):
    return v

def z(u, v):
    return u^3 - 3*u*v^2

def saddle_curve(u, v):
    return vector([x(u, v), y(u, v), z(u, v)])

def velocity_vector(u, v):
    return vector([diff(x(u, v), u), diff(y(u, v), u), diff(z(u, v), u)])

def frenet_frame(u, v):
    T = velocity_vector(u, v).normalized()  # Vecteur tangent unitaire
    a = vector([diff(T[i], u) for i in range(3)]).simplify_full()
    N = a.normalized() if a.norm() != 0 else vector([0, 0, 0])  # Vecteur normal unitaire
    B = T.cross_product(N)  # Vecteur binormal
    return T, N, B

r = 0.2
theta = var('theta')

def tube_surface(u, theta):
    T, N, B = frenet_frame(u, 0)
    return saddle_curve(u, 0) + r * cos(theta) * N + r * sin(theta) * B

plot_tube_saddle = parametric_plot3d((tube_surface(u, theta)[0], tube_surface(u, theta)[1], tube_surface(u, theta)[2]), 
                                     (u, -2, 2), (theta, 0, 2*pi), color='green', opacity=0.6, mesh=True)
show(plot_tube_saddle)
```
