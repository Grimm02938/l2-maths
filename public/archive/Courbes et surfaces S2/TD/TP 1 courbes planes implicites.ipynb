> # Courbes planes implicites

> Les courbes planes implicites sont données comme ensembles de solutions d'équations,
> ou comme courbes de niveau de fonctions de deux variables.

> ## Plusieurs façons de définir des fonctions ou expressions

> On souhaite tracer l'ensemble $C$ des points $(x, y)$ de $\R^2$ pour lesquels $x^2 + y^2 = 1$.
>
> - Si on note $E$ l'équation $x^2 + y^2 = 1$, on peut voir $C$ comme l'ensemble des solutions de $E$.
>
> - On peut voir $C$ comme l'ensemble de niveau $1$ de la fonction $f: \R^2 \to \R$ définie par $f(x, y) = x^2 + y^2$.
>
>   Cela s'écrit $\quad C = \{\,(x, y) \in \R^2\;|\; f(x, y) = 1\,\}\quad$ ou encore $\quad C = f^{-1}(\{1\})$.
>
> - L'équation $E$ peut aussi s'écrire sous la forme $x^2 + y^2 - 1 = 0$.
>
>   Donc $C$ est aussi l'ensemble de niveau $0$ de la fonction $g: \R^2 \to \R$ définie par $g(x, y) = x^2 + y^2 - 1$.
>
>   Cela s'écrit $\quad C = \{\,(x, y) \in \R^2\;|\; g(x, y) = 0\,\}\quad$ ou encore $\quad C = g^{-1}(\{0\})$.
>
> Dans Sage, la fonction `implicit_plot` permet de tracer des ensembles définis par des équations,
> ou définis comme ensembles de niveau $0$ d'une fonction.

> Voici plusieurs définitions d'objets qui en Sage peuvent représenter la fonction $g$ vue plus haut,
> celle qui au couple $(x, y)$ associe $x^2 + y^2 - 1$.

> - `p` est une fonction Python définie à l'aide d'une «fonction lambda»

```python
p = lambda x, y: x^2 + y^2 - 1
```

> ## Tracé de courbe



> On peut tracer une courbe à partir de tous ces objets:
> la courbe qui comprend les points $(x, y)$ tels que $g(x, y) = 0$.
>
> Les définitions de `p`, `q`, `r`, `s`, `t`, ne fournissent pas
> explicitement de points.
>
> Elles permettent seulement, quand on leur fournit un point, de décider s'il est solution ou non.
>
> On doit donc préciser dans quelle plage de valeurs de `x` et de `y` on veut faire le tracé.
>
> Le tracé est fait en échantillonnant des points disposés en réseau
> dans la zone de tracé: on repère à chaque point si la quantité $g(x, y)$
> est positive, négative, ou nulle, et on suppose que la courbe (où $g$ s'annule)
> sépare les zones où $g$ est positive des zones où $g$ est négative.



```python
trace_p = implicit_plot(p, (-1.125, 1.125), (-1.125, 1.125))
trace_p.show(figsize=3)
```

```python
trace_q = implicit_plot(q, (-1.125, 1.125), (-1.125, 1.125))
trace_q.show(figsize=3)
```

```python
trace_r = implicit_plot(r, (-1.125, 1.125), (-1.125, 1.125))
trace_r.show(figsize=3)
```

```python
trace_s = implicit_plot(s, (x, -1.125, 1.125), (y, -1.125, 1.125))
trace_s.show(figsize=3)
```

```python
trace_t = implicit_plot(t, (x, -1.125, 1.125), (y, -2, 2))
trace_t.show( aspect_ratio=1)
```

> On voit que les tracés sont essentiellement identiques:

```python
G = graphics_array([trace_p, trace_q, trace_r, trace_s, trace_t], nrows=1)
G.show(figsize=12)
```

> ## Calcul du gradient

> Le gradient est le vecteur des dérivées partielles.

> Dans Sage, aussi bien `r` que nous avons définie comme fonction,
> que `s` que nous avons définie comme expression symbolique,
> ont une méthode `gradient`.
>
> Cette méthode prend en argument un tuple de variables
> selon lesquelles calculer les dérivées partielles.
>
> En indiquant de dériver par rapport aux variables `(x, y)`, cela donne:

```python
grad_r = r.gradient((x, y))
grad_r
```

```python
grad_s = s.gradient((x, y))
grad_s
```

> Pour évaluer ce gradient,
>
> - dans le cas de `r`, on a obtenu une fonction qui à un couple `(x, y)`
>   associe le gradient de $g$ en ce point
> - dans le cas de `s`, on a obtenu une expression symbolique où on peut
>   remplacer `x` et `y` par des valeurs de notre choix
>   pour obtenir le gradient en un point choisi.

> Intéressons-nous au point $M$ de coordonnées $(a, b) = (3/5, 4/5)$.

```python
(a, b) = M = vector((3/5, 4/5))
```

```python
a, b
```

```python
M
```

> Calculons son gradient de plusieurs façons.

```python
grad_r_ab = grad_r(a, b)
grad_r_ab


```

```python
grad_r_M = grad_r(*M)
grad_r_M
```

```python
grad_s_ab = grad_s(x=a, y=b)
grad_s_ab
```

```python
grad_s_M = grad_s(x=M[0], y=M[1])
grad_s_M
```

> Le gradient est renvoyé sous forme de vecteur.

```python
parent(grad_r_ab), parent(grad_r_M), parent(grad_s_ab), parent(grad_s_M)
```

> ## Vecteur tangent

> Comme le gradient en un point d'une courbe de niveau est un vecteur normal à cette courbe,
> on peut trouver un vecteur tangent à l'aide d'une rotation d'un angle droit.
>
> Une telle rotation envoie un vecteur $(x, y)$ sur le vecteur $(-y, x)$
> si on tourne dans le sens trigonométrique (et son opposé si on tourne
> dans l'autre sens).
>
> On peut soit faire cette rotation à la main, soit en multipliant par une
> matrice de rotation.

> Rotation «à la main»:

```python
vecteur_normal_en_M = grad_s_M
nx, ny = vecteur_normal_en_M
vecteur_tangent_en_M = vector((-ny, nx))
vecteur_tangent_en_M
```

> Rotation par multiplication par une matrice de rotation:

```python
quart_de_tour = matrix([[0, -1], [1, 0]])
quart_de_tour
```

```python
vecteur_tangent_en_M = quart_de_tour * vecteur_normal_en_M
vecteur_tangent_en_M
```

>
> Préférez-vous la méthode à la main ou la méthode avec matrice?

```python
# Les deux méthodes sont compréhensibles mais la plus rapide et efficace est celle avec la matrice. 
```

> ## Annulation des dérivées partielles

> Rappelons que nous avons calculé les dérivées partielles, qui sont rassemblées dans le gradient.

```python
s_x
```

```python
s_y
```

```python
grad_s
```

> On sait que
>
> - pour que la courbe admette une tangente horizontale,
>   il faut que la dérivée partielle par rapport à $x$ s'annule
> - pour que la courbe admette une tangente verticale,
>   il faut que la dérivée partielle par rapport à $y$ s'annule


> Cherchons donc les points d'annulation des dérivées partielles le long de la courbe.
>
> Pour cela il faut résoudre un système d'équations:
>
> - une équation qui indique qu'on est sur la courbe voulue
> - une équation qui indique que la dérivée partielle par rapport à x s'annule

> On demande à Sage de se charger de la résolution avec la fonction `solve`.
> Cette fonction
>
> - prend en arguments une liste d'équations et une liste de variables
>   par rapport auxquelles on veut résoudre
> - renvoie une liste de solutions présentées sous forme de
>   listes d'équations donnant les valeurs des variables

```python
equations_pour_etre_sur_C_avec_derivee_x_nulle = [s == 0, s_x == 0]
variables = [x, y]
```

```python
points_de_C_avec_derivee_x_nulle = solve(equations_pour_etre_sur_C_avec_derivee_x_nulle, variables)
points_de_C_avec_derivee_x_nulle
```

> Les solutions sont donc ici le point $(0, -1)$ et le point $(0, 1)$.

```python
d(x,y) = (x^2 - y^2)^2 - 1
```

> Vérifiez que le point $(\sqrt{2}, 1)$ est solution de $H$.



```python
h(sqrt(2),1) #le point (sqrt(2),1) appartient à H car il vérifie son équation 
```

> Tracez le vecteur normal à $C_H$ en ce point.


```python
grad_h = h.gradient((x, y))
grad_h
```

```python
v = grad_h(sqrt(2),1)
```

```python
trace_h_grad = implicit_plot(h, (-10, 10), (-10, 10))+v.plot(start=(sqrt(2),1))
trace_h_grad.show(figsize=5)
```

> Déterminez l'équation de la tangente à $C_H$ en ce point.



```python
V = vector((sqrt(2),1))

quart_de_tour = matrix([[0, -1], [1, 0]])
vecteur_tangent = quart_de_tour * v
trace_droite_tangente = line([V - 0.7*vecteur_tangent, V+ 0.5*vecteur_tangent], color = 'red')

graph2 = vecteur_tangent.plot(start=[sqrt(2),1],color ='orange')+ trace_h_grad + trace_h + trace_droite_tangente + point2d([(sqrt(2), 1)], size = 50, color ='black')
graph2.show(figsize = 6)
```

```python
dx = diff(h,x)
dx
```

```python
solve([h==0, dx==0],[x,y])
```

```python
dy = diff(h,y)
dy
```

```python
solve([h==0, dy==0],[x,y])
```

> Donnez les équations de ces tangentes.

```python
def equation_tangentes(xp, yp):
    return(x - xp)*dx(xp, yp) + (y-yp)*dy(xp, yp) == 0

print(" équation (E1): ", equation_tangentes(0, 1))
print(" équation (E2): ", equation_tangentes(0, -1))
print(" équation (E3): ", equation_tangentes(1, 0))
print(" équation (E4): ", equation_tangentes(-1, 0))
```
