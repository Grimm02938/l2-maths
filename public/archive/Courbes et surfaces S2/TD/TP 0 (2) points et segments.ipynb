### Tuples

```python
p = (-1, 2)
q = (2, 3,)
```

```python
p
```

```python
(1)
```

## Tracé de points et de segments

La fonction `point2d` prend une liste de points, donnés comme
paires de coordonnées (sous forme de listes, tuples ou vecteurs),
et renvoie un objet graphique représentant ces points.

```python
Pab = point2d([a, b])
```

```python
Pab.show()
```

La fonction `line2d` prend une liste de points, donnés comme
paires de coordonnées (sous forme de listes, tuples ou vecteurs),
et renvoie un objet graphique représentant une ligne brisée
reliant ces points.

```python
Lab = line2d([a, b])
```

```python
Lab.show()
```

On peut superposer des figures en les additionnant.

```python
PLab = Pab + Lab
```

```python
PLab.show()
```

On peut préciser

- la couleur du tracé avec `color`,
- les valeurs minimales et maximales de x et y à prendre en compte avec `xmin`, `ymin`, `xmax`, `ymax`,

et on peut exiger la même échelle en `x` et en `y` avec `aspect_ratio`.

```python
Ppq = point2d([p, q], color='red')
Lpq = line2d([p, q], color='red')
PLpq = Ppq + Lpq
```

```python
PLpq.show(xmin=-5, xmax=5, ymin=-5, ymax=5, aspect_ratio=1)
```

### Mosaique

On peut créer une mosaïque de figures avec `graphics_array`, pour lequel on peut préciser le nombre de lignes avec `nrows` ou le nombre de colonnes avec `ncolumns`.

On peut préciser la taille de visualisation avec `figsize`.

```python
G = graphics_array([PLab, PLpq, PLab + PLpq], ncols=2)
```

```python
G.show(aspect_ratio=1, figsize=(8, 8))
```

## Réutiliser des options graphiques

On peut stocker des options graphiques dans un dictionnaire,
et les réutiliser pour plusieurs tracés.

```python
opt = dict(aspect_ratio=1, xmin=-5, ymin=-4, xmax=5, ymax=4)
```

```python
pts = [a, b]
opt_ab = dict(opt)
opt_ab['color'] = 'blue'
S_ab = point2d(pts, **opt_ab) + line2d(pts, **opt_ab)

pts = [p, q]
opt_pq = dict(opt)
opt_pq['color'] = 'red'
S_pq = point2d(pts, **opt_pq) + line2d(pts, **opt_pq)

pts = [u, v]
opt_uv = dict(opt)
opt_uv['color'] = 'green'
S_uv = point2d(pts, **opt_uv) + line2d(pts, **opt_uv)
```

```python
S = S_ab + S_pq + S_uv
```

```python
G = graphics_array([S_ab, S_pq, S_uv, S], ncols=2)
```

```python
G.show()
```
